## Data Combining

### "Table Bridge" creation

```python
GH_Bridge = GH[["ORDER_REF", "MERCHANT_ACCOUNT", "BATCH_NUMBER"]].groupby("ORDER_REF", as_index=False).first()
GH_Bridge.rename(columns = {"MERCHANT_ACCOUNT":"gh_MERCHANT_ACCOUNT", "BATCH_NUMBER":"gh_BATCH_NUMBER"}, inplace=True)

GH_Bridge
```

### "Transaction Lines" Table enriched by "Transactions" Table

```python
TL_Merged = tl.merge(
                    transactions,
                    on='TRANSACTION_ID',
                    how='left'
                )

TL_Merged
```

### Memory Management

#### <font color=gray>TL_Merged Compressing</font>

```python
gc.enable()
```

```python
gc.collect()
```

### Table Prepairing for Final GH Table Join

#### Excluding irrelevant Accounts ID

```python
S_Table = TL_Merged.loc[TL_Merged["ACCOUNT_ID"] != 128]
S_Table
```

#### Filtering _null_ values in "BATCH_NUMBER" column

```python
S_Table_F = pd.DataFrame(S_Table)

nan_values = S_Table_F[S_Table_F['BATCH_NUMBER'].isna()]

nan_values
```

### JOIN GH Table

#### Join "GH_Bridge" Table to a final subset


```python
S_JOIN = nan_values.merge(
                    GH_Bridge,
                    on='ORDER_REF',
                    how='left'
                )

S_JOIN
```
